```python
# Use a pipeline as a high-level helper
from transformers import pipeline
pipe = pipeline("text-classification", model="ProsusAI/finbert")
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
path = "/content/drive/MyDrive/FinBert/combined_clean_news.csv"
df = pd.read_csv(path)
df = df[700866:]
```

```python
title = df['title'].to_list()
print(len(title))
```

```python
import numpy as np

results = []

batch_size = 1000
for i in range(0, len(title), batch_size):
    prediction = pipe(title[i: i + batch_size])
    results.extend([d['label'] for d in prediction])

    if (i + batch_size) % 10000 == 0:
        print(len(results))
        np.save('/content/drive/MyDrive/FinBert/sentiment2.npy', np.array(results))

    print(f"Proccessed: {i + batch_size}")
```

```python
np.save('/content/drive/MyDrive/FinBert/sentiment2-2.npy', np.array(results))
print(len(results))
```

```python
df['sentiment'] = results
```
